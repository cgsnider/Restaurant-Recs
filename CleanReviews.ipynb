```python
import pandas as pd
import json
```

```python
##IMPORTANT: only run this cell if you are missing data/cleaned/yelp_academic_dataset_review.json
## Runtime ~4.5 minutes and requires >32GB of memory/swap space
business_df = pd.read_json('data/cleaned/yelp_academic_dataset_restaurant.json')

cols = ["review_id", "user_id", "business_id", "stars"]
data = []
file_name = 'data/yelp_reviews/yelp_academic_dataset_review.json'

with open(file_name, encoding='latin-1') as f:
    for line in f:
        doc = json.loads(line)
        lst = [doc[col] for col in cols]
        data.append(lst)

review_df = pd.DataFrame(data=data, columns=cols)


review_df = review_df[review_df['business_id'].isin(business_df['business_id'])]
review_df.to_json("data/cleaned/yelp_academic_dataset_review.json")
```
